## Part 0 : Importing libraries and dataset

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, confusion_matrix
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
%matplotlib inline
```

## Part 1 : Importing bagging classifier

Importing bagging classifier from sklearn.

It is important to note that we are not importing BAGGING REGRESSOR, we are importing BAGGING CLASSIFIER as this is a classification problem (We are require to predict the quality of wine).


```python
from sklearn.ensemble import BaggingClassifier
```

At the code below we have delibrately dropped out the split size hyperparameter. However by default it splits 75:25 ration

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
bag_dt = BaggingClassifier(DecisionTreeClassifier(), n_estimators = 10)

#BaggingClassifier(Name_the_ml_algo_to_be_used, number_of_ml_models_deployed)
```

```python
bag_dt.fit(x_train, y_train)
```

```python
bag_dt.predict(x_test)
```

```python
bag_dt.base_estimator
```

```python
bag_dt.classes_
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
bag_KNN = BaggingClassifier(KNeighborsClassifier(6), n_estimators = 10)

#BaggingClassifier(Name_the_ml_algo_to_be_used, number_of_ml_models_deployed)
```

```python
bag_KNN.fit(x_train, y_train)
```

```python
bag_KNN.base_estimator
```

```python
bag_KNN.classes_
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 5)
```

```python
rf.fit(x_train, y_train)
```

It is not neccesary that random forest will give the best accuracy which is the case below. It all depends on the NATURE of dataset and curatation of hyperparameters of a machine learning algorithum (KNN, RandomForest etc)

```python
rf.score(x_test, y_test)
```

```python
bag_dt.score(x_test, y_test)
```

```python
bag_KNN.score(x_test, y_test)
```

To view all the estimator, we will use below code

```python
rf.estimators_
```

In case we only want to view one of the estimator, we can use this code

```python
rf.estimators_[1]
```

Below the code to view one of the tree of our estimator. Here estimator is Random forest.

```python
plt.figure(figsize = (20, 10))
tree.plot_tree(rf.estimators_[1], filled = True)

# filled = True : will give you colored tree
```

Below is the code for viewing all the tree which was used for the mode

```python
plt.figure(figsize = (20, 20))

for i in range(len(rf.estimators_)):
    tree.plot_tree(rf.estimators_[i], filled = True)
```

## Note


It is important to note that Randomforest classifier is difficult to track but not with respect to the layout / design of the decison tree the design or decision incoporated within the tree. Rather it is difficult to track the bagging rules on which the sub-dataset was designed. However even this can also be know once the model is executed with the help of plotting the tree (reverse enginnering). 

```python
grid_param = {
    
    'n_estimators' : [5, 10, 15, 20, 15, 50, 75, 100, 125, 150],
    'criterion' : ["gini", "entropy", "log_loss"],
    'max_depth' : range(10),
    'min_samples_leaf' : range(10)
    
}
```

```python
rf = RandomForestClassifier()
```

```python
grid_search_rf = GridSearchCV(estimator = rf, param_grid = grid_param, cv = 7)
```

```python
grid_search_rf.fit(x_train, y_train)
```
